```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('migracion.csv')
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
col_numericas = df.select_dtypes(include=np.number).columns
for col in col_numericas:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1

  li = Q1 - 1.5 * IQR
  ls = Q3 + 1.5 * IQR

  df_filtrado = df[(df[col] >= li) & (df[col] <= ls)].copy()

df_filtrado
```

```python
df_filtrado['Razon_Migracion'] = df_filtrado['Razon_Migracion'].replace({'Económica': 'Trabajo', 'Conflicto': 'Guerra'})
df_filtrado
```

```python
np.mean(df_filtrado['Cantidad_Migrantes'])
print(f'La media de la columna Cantidad_Migrantes es: {np.mean(df["Cantidad_Migrantes"])}')
```

```python
np.median(df_filtrado['Cantidad_Migrantes'])
print(f'La mediana de la columna Cantidad_Migrantes es: {np.median(df["Cantidad_Migrantes"])}')
```

```python
PIB_promedio_origen = np.mean(df_filtrado['PIB_Origen'])
PIB_promedio_destino = np.mean(df_filtrado['PIB_Destino'])

print(f'El PIB promedio de los países de origen es: {PIB_promedio_origen}')
print(f'El PIB promedio de los países de destino es: {PIB_promedio_destino}')
```

```python
df_agrupado_razon = df_filtrado['Razon_Migracion'].value_counts()
df_agrupado_razon
```

```python
df_agrupado_razon_migrantes = df_filtrado.groupby('Razon_Migracion')['Cantidad_Migrantes'].sum()
df_agrupado_razon_migrantes
```

```python
promedio_IDH_origen = df_filtrado.groupby('Razon_Migracion')['IDH_Origen'].mean()
promedio_IDH_origen
```

```python
df_filtrado_ordenado = df_filtrado.sort_values(by='Cantidad_Migrantes', ascending=False)
df_filtrado_ordenado
```

```python
df_filtrado_ordenado_conflicto = df_filtrado_ordenado[df_filtrado_ordenado['Razon_Migracion'] == 'Guerra']
df_filtrado_ordenado_conflicto
```

```python
df_filtrado_ordenado_IDH_Destino = df_filtrado_ordenado[df_filtrado_ordenado['IDH_Destino'] > 0.90]
df_filtrado_ordenado_IDH_Destino
```

```python
df_filtrado_ordenado['Diferencia_IDH'] = df_filtrado_ordenado['IDH_Destino'] - df_filtrado_ordenado['IDH_Origen']
df_filtrado_ordenado
```

```python
df_filtrado_ordenado.to_csv('Migracion_Limpio.csv', index = False)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ventas.csv')
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df_limpio = df.dropna()
df_limpio
```

```python
df_limpio.duplicated()
```

```python
df_sin_duplicados = df_limpio.drop_duplicates()
df_sin_duplicados
```

```python
Q1 = df_sin_duplicados['Cantidad'].quantile(0.25)
Q3 = df_sin_duplicados['Cantidad'].quantile(0.75)
print('Q1 =', Q1)
print('Q3 =', Q3)

IQR = Q3 - Q1
print('IQR =', IQR)
```

```python
df_filtrado = df_sin_duplicados[(df_sin_duplicados['Cantidad'] >= Q1 - 1.5*IQR) & (df_sin_duplicados['Cantidad'] <= Q3 + 1.5*IQR)]
df_filtrado
```

```python
df_filtrado = df_filtrado.reset_index()
df_filtrado
```

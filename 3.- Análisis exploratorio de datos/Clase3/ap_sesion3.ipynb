```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
horas_estudio = np.random.randint(1, 20, 10)
calificacion_examen = 1.0 + (horas_estudio * 0.5) + np.random.uniform(-1, 1, 10 )
calificacion_examen[calificacion_examen < 1.0] = 1.0
calificacion_examen[calificacion_examen > 7.0] = 7.0

print("Horas de Estudio:", horas_estudio)
print("Calificación del Examen:", calificacion_examen)
```

```python
df = pd.DataFrame({
    'Ciudad de Residencia': ['Santiago', 'Valparaiso', 'Concepción'],
    'Preferencia de Transporte': ['Transporte Público', 'Bicicleta', 'Coche']
})

tabla_contingencia = pd.crosstab(df['Ciudad de Residencia'], df['Preferencia de Transporte'])
tabla_contingencia
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(horas_estudio, calificacion_examen)
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación del Examen')
plt.title('Relación entre Horas de Estudio y Calificación del Examen')
plt.grid(True)
plt.show()
```

```python
corr_matrix = np.corrcoef(horas_estudio, calificacion_examen)
corr_value = corr_matrix[0, 1]
print("Coeficiente de Correlación de Pearson:", corr_value)
```

Existe una correlación positiva fuerte entre las horas de estudio y la calificación obtenida en el examen, ya que el Coeficiente de Correlación de Pearson obtenido esta dentro del rango 0.8 y 1.0.

La correlación por sí sola no indica causalidad. Podrían existir otras variables correlacionadas con las horas de estudio (ej. la motivación del estudiante, el apoyo familiar, la calidad de la enseñanza, etc.) que también influyen en la calificación

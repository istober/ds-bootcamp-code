```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datos_trafico.csv')
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Día'], df['Vehículos'], color='red', marker='o', linestyle='--', linewidth=2, label='Vehículos por día')
plt.title('Vehículos Diario en Av. Principal')
plt.xlabel('Día')
plt.ylabel('Cantidad de Vehículos')
plt.xticks(df['Día'], rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Velocidad_Promedio'], bins=8, edgecolor='yellow')
plt.title('Distribución de Velocidad Promedio (Km/H)')
plt.xlabel('Velocidad Promedio (Km/H)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.scatter(df['Vehículos'], df['Accidentes'], color='brown', label='Accidentes diarios', marker='x')
plt.title('Relación entre Cantidad de Vehículos y Accidentes')
plt.xlabel('Cantidad de Vehículos')
plt.ylabel('Cantidad de Accidentes')
plt.grid(True)
plt.legend()
plt.savefig('grafico_accidentes.png')
plt.show()
```

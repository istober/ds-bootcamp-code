```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)
años_experiencia = np.random.randint(0, 10, 20)
salario = 500 + 250 * años_experiencia + np.random.normal(0, 50, 20)

print("Años de Experiencia:")
print(años_experiencia)
print("\nSalario (en dólares):")
print(salario)
```

```python
años_experiencia_r = años_experiencia.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(años_experiencia_r, salario)
```

```python
b_0 = modelo.intercept_
b_1 = modelo.coef_[0]

print(f'Intercepto: {b_0}')
print(f'Pendiente: {b_1}')
```

```python
salario_pred = modelo.predict(años_experiencia_r)

print("Primeros 5 salarios:")
print(np.round(salario_pred[:5]))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(salario, salario_pred)
mae = mean_absolute_error(salario, salario_pred)

print(f'Error cuadrático medio (MSE): {mse}')
print(f'Error Absoluto Medio (MAE): {mae}')
```

(MSE = 1522.99) Este número representa el promedio de las diferencias al cuadrado entre los salarios reales y los salarios predichos por el modelo. Como el MSE eleva las diferencias al cuadrado, este indicador no está en la misma unidad que el salario (está en dólares al cuadrado)
La raíz cuadrada del MSE, que sería aproximadamente 39.02, indica que, en promedio, las predicciones del modelo están a unos $39 del salario real, sugiriendo que hay algunos errores más grandes que están influyendo en el MSE más que en el MAE.

(MAE = 30.08) A diferencia del MSE, el MAE está en la misma unidad que la variable salario (en este caso dólares). Este MAE significa que, en promedio, las predicciones de salario del modelo están desviadas por aproximadamente $30 de los salarios reales.

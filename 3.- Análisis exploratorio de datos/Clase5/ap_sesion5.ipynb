```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = {
    'Altura_cm': [160, 175, 180, 165, 170],
    'Peso_kg': [55, 70, 75, 60, 65],
    'Edad': [20, 30, 25, 22, 28],
    'Horas_Ejercicio': [3, 5, 4, 2, 6]
}

df = pd.DataFrame(datos)
```

```python
corr_matrix = df.corr()
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5, linecolor='white')
plt.title('Matriz de Correlación')
plt.show()
```

Hay una correlación positiva perfecta (1.0) entre Altura_cm y Peso_kg. Esto sugiere que, a mayor altura, mayor es el peso.

Existe una correlación positiva fuerte (0.84) entre Edad y Horas_Ejercicio. Esto podría indicar que las personas de mayor edad se tiende a dedicar más horas al ejercicio.

Las otras variables tienen correlaciones positivas moderadas (0.5 a 0.69), lo que sugiere que hay alguna relación entre ellas, pero no tan fuerte como las anteriores. Por ejemplo, a mayor Edad, parece haber una tendencia a tener mayor Altura y Peso.

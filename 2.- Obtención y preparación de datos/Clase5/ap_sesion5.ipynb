```python
import pandas as pd
```

```python
datos = {
    'Estudiante': ['Juan', 'Juan', 'María', 'María'],
    'Materia': ['Matemáticas', 'Historia', 'Matemáticas', 'Historia'],
    'Calificación': [6.5, 5.8, 4.2, 6.0]
}
df = pd.DataFrame(datos)
df
```

```python
df.set_index(['Estudiante', 'Materia'], inplace=True)
df
```

```python
df.loc[('María', 'Historia'), :]
```

```python
df = pd.DataFrame(datos)
df.groupby('Materia')['Calificación'].mean()
```

```python
df.groupby('Materia')['Calificación'].max()
```

```python
df_pivot = df.pivot_table(values='Calificación', index='Estudiante', columns='Materia')
df_pivot
```

```python
df_pivot=df_pivot.reset_index()
df_pivot
```

```python
df_melted = pd.melt(df_pivot, id_vars='Estudiante', var_name='Materia', value_name='Calificación')
df_melted
```
